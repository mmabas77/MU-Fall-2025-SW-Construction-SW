## 1. Traditional Programming vs. Machine Learning

### Traditional Programming Approach

In traditional programming:
- **Input**: Data + Rules (written by programmers)
- **Output**: Answers

```
┌────────────┐     ┌────────────┐
│   Data     │────►│            │
└────────────┘     │  Computer  │────► Answers
┌────────────┐     │            │
│   Rules    │────►│            │
└────────────┘     └────────────┘
```

### Machine Learning Approach

In machine learning:
- **Input**: Data + Answers (labeled examples)
- **Output**: Rules (learned patterns)

```
┌────────────┐     ┌────────────┐
│   Data     │────►│            │
└────────────┘     │  Computer  │────► Rules
┌────────────┐     │            │
│  Answers   │────►│            │
└────────────┘     └────────────┘
```

**Key Insight**: Machine Learning allows algorithms to *learn rules from labeled examples* (training data). The learned rules should also be able to generalize to correctly recognize or predict new examples not in the training set.

---

## 2. What is Machine Learning?

> **Definition**: Machine Learning is the study of computer programs (algorithms) that can learn by example. ML algorithms can generalize from existing examples of a task.

For example, after seeing a training set of labeled images, an image classifier can figure out how to apply labels accurately to new, previously unseen images.

**Machine Learning brings together:**
- Statistics
- Computer Science
- Domain Knowledge
- And more...

### Pure ML vs. Applied ML

| Pure ML | Applied ML |
|---------|------------|
| Concerned with algorithm behavior as a whole | Concerned with behavior on specific real-life datasets |
| Theoretical foundations | Practical implementations |
| Algorithm development | Problem solving |

### 3.1 Supervised Learning

Learn to predict target values from **labeled data**.

**Classification** (target values are discrete classes):
- Fraud detection
- Email spam detection
- Medical diagnostics
- Image classification

**Regression** (target values are continuous values):
- Risk assessment
- Score prediction
- Price forecasting

### 3.2 Unsupervised Learning

Find structure in **unlabeled data**.

- **Clustering**: Find groups of similar instances in the data
- **Outlier Detection**: Finding unusual patterns

### 3.3 Reinforcement Learning

- Learn from interaction with an environment
- Agent receives rewards or penalties for actions
- Some algorithms can deal with partially labeled training data

## 8. Hands-On: Object Recognition System

Let's build a simple fruit classification system using the **Fruit Dataset**.

### Dataset Information

The fruit dataset contains measurements of various fruits with features like:
- Mass
- Width
- Height
- Color score

Source: [Fruit Dataset on GitHub](https://github.com/susanli2016/Machine-Learning-with-Python/blob/22c63e943461dab8acdd02bac45710959a90ef65/fruit_data_with_colors.txt)

````python
# Step 1: Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# For machine learning
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

print("Libraries imported successfully!")
````

````python
# Step 2: Load the dataset
# You can download the dataset from the GitHub link above
# For this example, we'll create a sample dataset

# Try to load from file, or create sample data
try:
    fruits = pd.read_table('fruit_data_with_colors.txt')
except FileNotFoundError:
    # Create sample fruit data if file not found
    np.random.seed(42)
    n_samples = 59
    
    data = {
        'fruit_label': np.random.choice([1, 2, 3, 4], n_samples),
        'fruit_name': ['apple', 'mandarin', 'orange', 'lemon'] * 15,
        'fruit_subtype': ['granny_smith', 'mandarin', 'orange', 'lemon'] * 15,
        'mass': np.random.randint(70, 250, n_samples),
        'width': np.random.uniform(5.5, 9.5, n_samples).round(1),
        'height': np.random.uniform(4.0, 10.0, n_samples).round(1),
        'color_score': np.random.uniform(0.45, 0.95, n_samples).round(2)
    }
    data['fruit_name'] = data['fruit_name'][:n_samples]
    data['fruit_subtype'] = data['fruit_subtype'][:n_samples]
    fruits = pd.DataFrame(data)
    print("Using sample dataset")

# Display the first few rows
print("Dataset Shape:", fruits.shape)
fruits.head()
````

````python
# Step 3: Explore the dataset
print("Dataset Information:")
print("=" * 40)
print(f"Number of samples: {len(fruits)}")
print(f"Number of features: {len(fruits.columns) - 3}")  # Excluding label and name columns
print("\nFeature columns:")
print(fruits[['mass', 'width', 'height', 'color_score']].describe())
````

````python
# Step 4: Check class distribution
print("Class Distribution:")
print(fruits['fruit_label'].value_counts())

# Visualize class distribution
plt.figure(figsize=(8, 5))
fruits['fruit_label'].value_counts().plot(kind='bar', color=['red', 'orange', 'darkorange', 'yellow'])
plt.title('Distribution of Fruit Classes')
plt.xlabel('Fruit Label')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
````

````python
# Step 5: Prepare features and target variable
feature_names = ['mass', 'width', 'height', 'color_score']

X = fruits[feature_names]  # Features
y = fruits['fruit_label']  # Target variable

print("Features (X) shape:", X.shape)
print("Target (y) shape:", y.shape)
````

````python
# Step 6: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.25,    # 25% for testing
    random_state=0     # For reproducibility
)

print(f"Training set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")
````

````python
# Step 7: Train a K-Nearest Neighbors classifier
knn = KNeighborsClassifier(n_neighbors=5)

# Fit the model on training data
knn.fit(X_train, y_train)

print("Model trained successfully!")
````

````python
# Step 8: Evaluate the model
train_accuracy = knn.score(X_train, y_train)
test_accuracy = knn.score(X_test, y_test)

print("Model Evaluation:")
print("=" * 40)
print(f"Training Accuracy: {train_accuracy:.2%}")
print(f"Test Accuracy: {test_accuracy:.2%}")
````

````python
# Step 9: Make predictions on new data
# Let's predict the fruit type for a new sample

# Example: A fruit with mass=150g, width=7.5cm, height=7.0cm, color_score=0.75
new_fruit = [[150, 7.5, 7.0, 0.75]]

prediction = knn.predict(new_fruit)
fruit_names = {1: 'apple', 2: 'mandarin', 3: 'orange', 4: 'lemon'}

print(f"Prediction for new fruit: {fruit_names.get(prediction[0], 'unknown')}")
````

````python
# Step 10: Visualize the data (2D projection using height and width)
plt.figure(figsize=(10, 6))

# Create a scatter plot
colors = ['red', 'orange', 'darkorange', 'yellow']
labels = ['Apple', 'Mandarin', 'Orange', 'Lemon']

for fruit_label in fruits['fruit_label'].unique():
    subset = fruits[fruits['fruit_label'] == fruit_label]
    plt.scatter(
        subset['height'], 
        subset['width'],
        c=colors[fruit_label-1],
        label=labels[fruit_label-1],
        s=100,
        alpha=0.7,
        edgecolors='black'
    )

plt.xlabel('Height (cm)')
plt.ylabel('Width (cm)')
plt.title('Fruit Classification: Height vs Width')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````
