## Part 1: Introduction to Machine Learning

### 1.1 What is Machine Learning?

Machine Learning is the study of computer programs (algorithms) that can **learn by example**. ML algorithms can generalize from existing examples of a task.

For example, after seeing a training set of labeled images, an image classifier can figure out how to apply labels accurately to new, previously unseen images.

> **Key Insight:** Machine Learning brings together **statistics**, **computer science**, and more!

Learn to predict target values from **labeled data**.

| Type | Description |

Learning through interaction with an environment, receiving rewards or penalties for actions taken.

## Part 2: Python for Machine Learning

### 2.1 Why Python?

Python was invented in the late 1980s and has become the dominant language for Machine Learning because of:

1. **Readable and Maintainable Code** - Clean syntax helps maintain and update software efficiently
2. **Cross-Platform Compatibility** - Works on all major operating systems
3. **Multiple Programming Paradigms** - Supports object-oriented, structured, and functional programming
4. **Dynamic Typing** - Interpreter assigns types at runtime
5. **Rich Ecosystem** - Powerful libraries for data science (NumPy, Pandas, Matplotlib, scikit-learn)

### 2.2 Python Key Features

- Uses **indentation** as a syntactic feature
- Interpreted language
- Object-oriented (with encapsulation support)

  - `__variable` â†’ private variable

### 2.3 Modules vs Packages

| Modules | Packages |

## Part 3: NumPy - Numerical Python

NumPy provides fundamental data structures used by scikit-learn, particularly **multi-dimensional arrays**. Data input to scikit-learn is typically in the form of a NumPy array.

````python
# Import NumPy
import numpy as np
````

### 3.1 Creating Arrays

````python
# Creating an array from a Python list
my_list = [0, 1, 2, 3, 4]
arr = np.array(my_list)
print("Array from list:", arr)
````

````python
# Using np.arange() - similar to Python's range()
print("arange(0,10):     ", np.arange(0, 10))
print("arange(0,10,3):   ", np.arange(0, 10, 3))  # with step of 3
````

````python
# Creating arrays filled with zeros and ones
print("5x5 zeros matrix:")
print(np.zeros((5, 5)))

print("\n2x4 ones matrix:")
print(np.ones((2, 4)))
````

````python
# Random integers
print("Single random int (0-9):  ", np.random.randint(0, 10))
print("\n3x3 random int matrix (0-99):")
print(np.random.randint(0, 100, (3, 3)))
````

````python
# np.linspace() - evenly spaced numbers over a specified interval
print("6 evenly spaced values from 0 to 10:")
print(np.linspace(0, 10, 6))
````

### 3.2 Array Operations

````python
# Setting a random seed for reproducibility
np.random.seed(101)
arr2 = np.random.randint(0, 100, 10)
print("Array:", arr2)
````

````python
# Basic statistical operations
print("Maximum value:", arr2.max())
print("Minimum value:", arr2.min())
print("Mean value:   ", arr2.mean())
print("Index of min: ", arr2.argmin())
print("Index of max: ", arr2.argmax())
````

````python
# Reshaping arrays
print("Original array shape:", arr2.shape)
print("\nReshaped to 2x5:")
print(arr2.reshape(2, 5))
````

### 3.3 Array Indexing and Slicing

````python
# Create a 10x10 matrix
mat = np.arange(0, 100).reshape(10, 10)
print("10x10 Matrix:")
print(mat)
````

````python
# Accessing single elements
row = 0
col = 1
print(f"Element at row {row}, col {col}:", mat[row, col])
````

````python
# Slicing - getting entire rows or columns
print("First column (all rows, column 0):")
print(mat[:, 0])

print("\nFirst row (row 0, all columns):")
print(mat[0, :])
````

````python
# Slicing sub-matrices
print("Top-left 3x3 submatrix:")
print(mat[0:3, 0:3])

print("\nUsing negative indices (rows 0-2, columns -3 to -1):")
print(mat[0:3, -3:-1])
````

## Part 4: Pandas - Data Analysis Library

Pandas provides key data structures like **DataFrame** for tabular data, along with support for reading/writing data in different formats.

````python
import pandas as pd
````

### 4.1 Pandas Series

A Series is a one-dimensional labeled array capable of holding any data type.

````python
# Creating a Series from a list
step_data = [3620, 7891, 9761, 3907, 4338, 5373]
step_counts = pd.Series(step_data, name='steps')
print("Original Series:")
print(step_counts)
````

````python
# Setting a custom index (date range)
step_counts.index = pd.date_range('20240329', periods=6)
print("Series with date index:")
print(step_counts)
````

````python
# Accessing data by label (like a dictionary)
print("Value at 2024-04-01:", step_counts['2024-04-01'])

# Accessing by index position (like an array)
print("Value at index 3:   ", step_counts[3])
````

````python
# Check data types
print("Data type:", step_counts.dtypes)
````

### 4.2 Pandas DataFrame

A DataFrame is a 2-dimensional labeled data structure with columns of potentially different types.

````python
# Creating a DataFrame from multiple lists
cycling_data = [10.7, 0, None, 2.4, 15.3, 10.9]

# Create a list of tuples
joined_data = list(zip(step_data, cycling_data))
print("Joined data as tuples:")
print(joined_data)
````

````python
# Create DataFrame with custom index and column names
activity_df = pd.DataFrame(
    joined_data,
    index=pd.date_range('20240329', periods=6),
    columns=['Walking', 'Cycling']
)
print("Activity DataFrame:")
print(activity_df)
````

````python
# Accessing a single column
print("Walking column:")
print(activity_df['Walking'])
````

### 4.3 DataFrame Indexing: loc vs iloc

| Method | Description |
|--------|-------------|
| `.loc` | Label-based indexing (uses actual labels/names) |
| `.iloc` | Integer-based indexing (uses positions) |

````python
# Using .loc (label-based)
print("Row at '2024-04-01':")
print(activity_df.loc['2024-04-01'])
````

````python
# Slicing with .loc (includes end point!)
print("Rows from '2024-04-01' to '2024-04-03':")
print(activity_df.loc['2024-04-01':'2024-04-03'])
````

````python
# Combining row and column selection with .loc
print("Walking data from '2024-04-01' to '2024-04-03':")
print(activity_df.loc['2024-04-01':'2024-04-03', ['Walking']])
````

````python
# Using .iloc (position-based)
print("First row (index 0):")
print(activity_df.iloc[0, :])

print("\nFirst two rows, first column:")
print(activity_df.iloc[:2, 0])
````

### 4.4 Working with Real Datasets

Let's work with the famous Iris dataset to practice data manipulation.

````python
# Load the Iris dataset from scikit-learn (for convenience)
from sklearn.datasets import load_iris

# Load and convert to DataFrame
iris = load_iris()
data = pd.DataFrame(iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
data['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

print("First 5 rows:")
data.head()
````

````python
# Basic DataFrame information
print("Shape (rows, columns):", data.shape)
print("\nColumn names:", list(data.columns))
````

````python
# Accessing columns
print("Sepal width column (first 10 values):")
print(data['sepal_width'].head(10))
````

````python
# Conditional filtering
print("Rows where sepal_width > 3.5:")
print(data.loc[data.sepal_width > 3.5].head())
````

````python
# Creating a new column
data['sepal_area'] = data.sepal_length * data.sepal_width
print("DataFrame with new 'sepal_area' column:")
print(data.iloc[:5, -3:])
````

````python
# Group by and count
print("Number of samples per species:")
print(data.groupby('species').size())
````

### 4.5 Statistical Measures

In Machine Learning (and mathematics), there are key statistical values:

1. **Mean** - The average value
2. **Median** - The middle value when sorted
3. **Mode** - The most common value
4. **Standard Deviation** - How spread out the values are
5. **Variance** - The average of squared differences from the mean

````python
# Central tendency measures
print("Sepal width - Mean:  ", data.sepal_width.mean())
print("Petal length - Median:", data.petal_length.median())
print("Petal length - Mode:  ", data.petal_length.mode().values)
````

````python
# Spread measures
print("Petal length statistics:")
print(f"  Standard Deviation: {data.petal_length.std():.4f}")
print(f"  Variance:           {data.petal_length.var():.4f}")
print(f"  Standard Error:     {data.petal_length.sem():.4f}")
````

Histograms show the distribution of a single variable.

Scatter plots show the relationship between two variables.
